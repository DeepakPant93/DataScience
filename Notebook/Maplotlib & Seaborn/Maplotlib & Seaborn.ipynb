```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
#create a line chart with the help of matplotlib 
```

```python
x = np.arange(1,9)
x
```

```python
y= np.array([2,4,8,12,16,18,20,22])
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y,color= "g", linestyle=":", linewidth=2)
plt.title("graph between x and y")
plt.xlabel("values of x")
plt.ylabel("value of y")
plt.show()
```

```python
z= [1,9,7,45,12,68,41,15]
```

```python
#plotting multiple lines

plt.plot(x,y,color= "g", linestyle=":", linewidth=2)
plt.plot(x,z,color= "r", linestyle="-.", linewidth=2)
plt.title("graph between x&y and y&z")
plt.xlabel("values of x")
plt.show()
```

```python
#to create grid line
#use plt.grid
```

```python
plt.plot(x,y,color= "g", linestyle=":", linewidth=2)
plt.title("graph between x and y")
plt.xlabel("values of x")
plt.ylabel("value of y")
plt.grid()
plt.show()
```

**#use "--"" for dashed line <br>
#use "-" for solid line <br>
#use "-." for dashdot line <br>
#use ":" for dotted line <br>

## Creating subplot

```python
plt.subplot(1,2,1)
plt.plot(x,y,color="g", linestyle=":", linewidth=2)
plt.ylabel("value of y")
plt.subplot(1,2,2)
plt.plot(x,z,color="r", linestyle=":", linewidth=2)
plt.ylabel("value of z")
plt.show()
```

### Bar plot

```python
performance={"julia":98,"andrew":78,"chris":75,"mathew":76}
```

```python
name= list(performance.keys())
marks= list(performance.values())
```

```python
plt.bar(name,marks)
plt.show()
```

```python
plt.bar(name, marks, color="m")
plt.title("marks of students")
plt.xlabel("marks")
plt.ylabel('names')
plt.grid()
plt.show()
```

```python
### Scatter plot
```

Scatter plots are used to plot data points on a horizontal and a vertical axis in the attempt to show how much one variable is affected by another. Each row in the data table is represented by a marker whose position depends on its values in the columns set on the X and Y axes.

```python
#lets's create two lists
```

```python
a1= [10,15,14,13,18,17]
a2=[12,45,78,14,13,11]
```

```python
plt.scatter(a1,a2)
plt.show()
```

```python
a1= [10,15,14,13,18,17]
a2=[12,45,78,14,13,11]
a3=[67,53,66,35,9,78]
```

```python
#plotting two scatter plots with a slight diffrent asthetics
plt.scatter(a1,a2,marker="*", color="r",s=100)
plt.scatter(a1,a3, color="g")
plt.title("random distribution")
plt.ylabel("a2 & a3")
plt.xlabel("a1")
plt.show()
```

```python
#plotting subplots
plt.subplot(1,2,1)
plt.scatter(a1,a2,marker="*", color="r",s=100)
plt.xlabel("a1")
plt.ylabel("a2")
plt.subplot(1,2,2)
plt.scatter(a1,a3, color="g")
plt.ylabel("a3")
plt.xlabel("a1")
plt.show()
```

### Box Plot

Also known as five number summary as it shows 0, 25, 50,75,100 percentile of a given data.

```python
a1=[10,20,30,40,50,60,70,150,152,153,149,160,170]
```

```python
plt.boxplot(a1)
plt.show()
```

```python
#plotting 3 box plots
a1=[10,20,30,40,50,60,70]
a2=[4,67,35,26,75,84,59]
a3=[67,53,66,35,9,78,23]
```

```python
data= list([a1,a2,a3])
```

```python
plt.boxplot(data)
plt.show()
```

```python
fruit=["apple","mango","peach","banana"]
quantity=[3,6,9,12]
```

```python
plt.pie(quantity,labels=fruit)
plt.show()
```

```python
#changing aesthetics of pie chart, we can use autopct for showing percentage
plt.pie(quantity, labels= fruit, autopct= '%0.1f%%', colors= ['red','yellow','cyan','green'], radius=1)
plt.show()
```

### donut chart

```python
plt.pie(quantity, labels= fruit, autopct= '%0.1f%%', colors= ['red','yellow','cyan','green'], radius=1)
plt.pie([10], colors="w", radius=0.4)
plt.show()
```
